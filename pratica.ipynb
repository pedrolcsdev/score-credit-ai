### Case: Score de Crédito dos Clientes
Você foi contratado por um banco para conseguir definir o score de crédito dos clientes. Você precisa analisar todos os clientes do banco e, com base nessa análise, criar um modelo que consiga ler as informações do cliente e dizer automaticamente o score de crédito dele: Ruim, Ok, Bom

### Bibliotecas Utilizadas
- Pandas  
  Utilizada para importar, organizar e manipular os dados da base de clientes.

- Scikit-learn (sklearn)  
  Biblioteca de Machine Learning utilizada para:
  - tratamento dos dados
  - separação entre treino e teste
  - criação dos modelos
  - avaliação dos resultados

### Modelos de Machine Learning Utilizados

#### - Random Forest Classifier
Modelo baseado em múltiplas árvores de decisão.  

#### - K-Nearest Neighbors (KNN)  

Ele compara novos clientes com outros clientes parecidos já existentes na base de dados.

#### Objetivo dos Modelos

Os modelos foram treinados utilizando dados históricos dos clientes para aprender padrões financeiros e, posteriormente, prever automaticamente o score de crédito de novos clientes.

```python
import pandas as pd

# Passo 1: importar a base
tabela = pd.read_csv("clientes.csv")

display(tabela)
display(tabela.info())
```

```python
from sklearn.preprocessing import LabelEncoder

# Passo 2: transformar colunas de texto em números
codificador_profissao = LabelEncoder()
tabela["profissao"] = codificador_profissao.fit_transform(tabela["profissao"])

codificador_mix = LabelEncoder()
tabela["mix_credito"] = codificador_mix.fit_transform(tabela["mix_credito"])

codificador_comportamento = LabelEncoder()
tabela["comportamento_pagamento"] = codificador_comportamento.fit_transform(tabela["comportamento_pagamento"])

display(tabela)
```

```python
# y = o que eu quero prever
y = tabela["score_credito"]

# x = informações usadas para fazer a previsão
x = tabela.drop(columns=["score_credito", "id_cliente"])
```

```python
# Passo 3: separar treino e teste
from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3, random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

# Passo 4: criar os modelos
modelo_arvoredecisao = RandomForestClassifier()
modelo_knn = KNeighborsClassifier()

# treinar os modelos
modelo_arvoredecisao.fit(x_treino, y_treino)
modelo_knn.fit(x_treino, y_treino)
```

```python
# Passo 5: avaliar os modelos
from sklearn.metrics import accuracy_score

previsao_arvore = modelo_arvoredecisao.predict(x_teste)
previsao_knn = modelo_knn.predict(x_teste)

print("Acurácia Árvore de Decisão:", accuracy_score(y_teste, previsao_arvore))
print("Acurácia KNN:", accuracy_score(y_teste, previsao_knn))
```

```python
# Passo 6: fazer novas previsões
novos_clientes = pd.read_csv("novos_clientes.csv")

novos_clientes["profissao"] = codificador_profissao.transform(novos_clientes["profissao"])
novos_clientes["mix_credito"] = codificador_mix.transform(novos_clientes["mix_credito"])
novos_clientes["comportamento_pagamento"] = codificador_comportamento.transform(novos_clientes["comportamento_pagamento"])

previsao = modelo_arvoredecisao.predict(novos_clientes)

display(novos_clientes)
display(previsao)
```
